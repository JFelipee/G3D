#### Texturas
Cuando dibujamos una figura, ya sea un triángulo, cuadrado, o polígono cualquiera, podemos rellenar en lugar de un colo plano, o una gama de colores, podemos hacerlo con una *textura*, esto es, con un patrón de diseño.

***Actividad*** obtener una imagien cuadrada de formato JPG.

```python
#Librerías
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

from PIL import Image 


#La varibale 'vertex' lleva guarda de los vértices del cuadrado que queremos dibujar
#La variable 'square' lleva guarda del cuadrado [0,1]x[0,1]
vertex= [[-1,-1], [1,-1],[1,1],[-1,1]]
square= [[0,0],[1,0],[1,1],[0,1]]


#Cargamos la imagen que hemos descargado con "Image.open()"
texture = Image.open("imagen.jpg")
xtexture= texture.size[0]
ytexture= texture.size[1]
#[xtexture,ytexture]= texture.size

texturebyte= texture.tobytes("raw", "RGBX",0,-1)

#función
def init():
    glClearColor(0.0,0.0,0.0,0.0)
    gluOrtho2D(-2.0,2.0,-2.0,2.0)
    
def dibujo():
    glClear(GL_COLOR_BUFFER_BIT)
    glTexParameterf(GL_TEXTURE_2D,GL_TEXTURE_MAG_FILTER, GL_NEAREST)
    glTexParameterf(GL_TEXTURE_2D,GL_TEXTURE_MIN_FILTER, GL_NEAREST)
    glTexImage2D(GL_TEXTURE_2D, 0, 3,xtexture, ytexture,0,GL_RGBA,GL_UNSIGNED_BYTE,
                 texturebyte)
    glEnable(GL_TEXTURE_2D)
    glBegin(GL_QUADS)
    for i in range(4):
        #intercalamos los vértices de la imagen con los vértices de nuestro objeto
        glTexCoord2fv(square[i])#coordenadas de imagen
        glVertex(vertex[i])#coordenadas de cuadrado
    glEnd()
    glDisable(GL_TEXTURE_2D)
    glFlush()
    
glutInit(sys.argv)
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowPosition(300,300)
glutInitWindowSize(500,500)
glutCreateWindow("Mi primera textura")
init()

glutDisplayFunc(dibujo)

glutMainLoop()
```

```python
#Un poco más allá de la textura

#Librerías
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

from PIL import Image 


#La varibale 'vertex' lleva guarda de los vértices del cuadrado que queremos dibujar
#La variable 'square' lleva guarda del cuadrado [0,1]x[0,1]
vertex= [[-1,-1], [1,-1],[1,1],[-1,1]]
square= [[0,0],[1,0],[1,1],[0,1]]


#Cargamos la imagen que hemos descargado con "Image.open()"
texture = Image.open("imagen.jpg")
xtexture= texture.size[0]
ytexture= texture.size[1]
#[xtexture,ytexture]= texture.size

texturebyte= texture.tobytes("raw", "RGBX",0,-1)

#función
def init():
    glClearColor(0.0,0.0,0.0,0.0)
    gluOrtho2D(-2.0,2.0,-2.0,2.0)
    
def dibujo():
    glClear(GL_COLOR_BUFFER_BIT)
    glTexParameterf(GL_TEXTURE_2D,GL_TEXTURE_MAG_FILTER, GL_NEAREST)
    glTexParameterf(GL_TEXTURE_2D,GL_TEXTURE_MIN_FILTER, GL_NEAREST)
    glTexImage2D(GL_TEXTURE_2D, 0, 3,xtexture, ytexture,0,GL_RGBA,GL_UNSIGNED_BYTE,
                 texturebyte)
    glEnable(GL_TEXTURE_2D)
    glEnableClientState(GL_VERTEX_ARRAY)
    glEnableClientState(GL_TEXTURE_COORD_ARRAY)
    glVertexPointer(2,GL_INT,0,vertex)
    glTexCoordPointer(2,GL_FLOAT,0,square)
    index= list(range(4))
    glDrawElements(GL_QUADS,len(index), GL_UNSIGNED_BYTE,index)
    glDisableClientState(GL_TEXTURE_COORD_ARRAY)
    glDisableClientState(GL_VERTEX_ARRAY)
    glDiable(GL_TEXTURE_2D)
    
    glFlush()
    
glutInit(sys.argv)
glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)
glutInitWindowPosition(300,300)
glutInitWindowSize(500,500)
glutCreateWindow("Mi primera textura")
init()

glutDisplayFunc(dibujo)

glutMainLoop()
```
