Ejemplos de:

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys
import numpy as np

def init():
    """
    Función que inicializa la ventana
    """
    #Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    #El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    #El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0,-2.0,2.0)
    
    
def trianguloE():
    glClear(GL_COLOR_BUFFER_BIT)
    
     
    # Definir el color que tendrán los objetos que vamos a dibujar
    glColor3f(1.0, 0.0, 0.0)
 
    glLineWidth(2.0)
    glBegin(GL_LINE_STRIP)
    lados=6
    for k in range(lados + 1): #para cerrar el hexagono ya que solo encadena y llega hasta el lado 6 y se necesita un 7 vértice
        glVertex(np.cos(2*np.pi/lados*k), np.sin(2*np.pi/lados*k))
    
    glEnd()
    glFlush()
    
glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Probando figuras")# Creamos la ventana

init()# Inicializamos aquí a la ventana

glutDisplayFunc(trianguloE)

glutMainLoop()
```

### Polígonos

La figura más básica para dibujar es el triángulo, ¿Por qué?
1. Primero porque es el que tiene menos lados.
2. Pero(más importante) es que todo polígono se puede subdividir en triángulos.

Los triángulos están definidos por sus tres vértices. Y en `OpenGL`... podemos dibujar un triángulo a través de sus tres vértices con
    
    glBegin(GL_TRIANGLES)
    #vertices
    glEnd()
El conjunto de vértices que le damos como dato los agrupa de tres en tres y fomar con cada grupo el triangulo que determinan esos tres vertices.

Si queremos realizar figura un poco más complejas a partir de pegados de triangulos, este comando posiblemente nos haria repetir una y otra vez los vertices. Para esto tenemos otra dos versiones: 
1. `GL_TRIANGLE_STRIP`: Si los vértices son P1, P2 ... Pn, realiza agrupaciones de tres en tres Pk, Pk+1, Pk+2--> k=1.. n-2

2. `GL_TRIANGLES_FAN` : Si los vértices son P0, P1 ... Pn, realiza argupaciones de tres eb tres P0, Pk, Pk+1

```python
#Para el primer ejemplo

from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys
import numpy as np

def init():
    """
    Función que inicializa la ventana
    """
    #Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    #El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    #El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0,-2.0,2.0)
    
    
def trianguloE():
    glClear(GL_COLOR_BUFFER_BIT)
    
     
    # Definir el color que tendrán los objetos que vamos a dibujar
    glColor3f(1.0, 0.0, 0.0)
 
    glLineWidth(2.0)
    glBegin(GL_TRIANGLE_STRIP)
    lados=6
    for k in range(lados): #para cerrar el hexagono ya que solo encadena y llega hasta el lado 6 y se necesita un 7 vértice
        glVertex(np.cos(2*np.pi/lados*k), np.sin(2*np.pi/lados*k))
    
    glEnd()
    glFlush()
    
glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Probando figuras")# Creamos la ventana

init()# Inicializamos aquí a la ventana

glutDisplayFunc(trianguloE)

glutMainLoop()
```

### Transparencias básicas 
Para pode añadir transparencias, tenemos que usar una cuarta coordenada que se conoce como "aplha", y va de 0 a 1 como el resto de las coordenadas: 0 es para "opaci" y 1 es para "transparente"

Luego en lugar de usar RGB, se tendría que:
* usar RGBA. "glutInitDisplayMode(GLUT_RGBA)"
* Usar colores con cuatro componentes "glColo4f()", donde la última coordenada es el *alpha*
* Habilitar las transparencias
    
    glEnable(GL_BLEND)
    
* y como se mezclan: esto se determina con la función : "glBlendFunc()" con 2 argumentos 
    
    glBlendFunc(GL_SRC_ALPHA, GL_ONE_MINUS_SRC_ALPHA)

* Se puede descactivar con enable
    
    glEnable


```python

from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys
import numpy as np

def init():
    """
    Función que inicializa la ventana
    """
    #Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    #El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    #El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0,-2.0,2.0)
    
    
def trianguloE():
    glClear(GL_COLOR_BUFFER_BIT)
    
    
    glLineWidth(2.0)
    
    glEnable(GL_BLEND)
    glBlendFunc(GL_SRC_ALPHA, GL_ONE_MINUS_SRC_ALPHA)
    
    glBegin(GL_POLYGON)
    glColor4f(0.0,0.0,1.0,0.25)#Rojo
    glVertex2f(0,-2)
    glVertex2f(1,1)
    glVertex2f(-1,1)
    
    glColor4f(1.0,0.0,0.0,1)#Azul
    glVertex2f(0,2)
    glVertex2f(1,-1)
    glVertex2f(-1,-1)

    
    glEnd()
    glDisable(GL_BLEND)
    glFlush()
    
glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGBA)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Probando figuras")# Creamos la ventana

init()# Inicializamos aquí a la ventana

glutDisplayFunc(trianguloE)

glutMainLoop()
```

## Crear Objetos

Hay varias formas de crear objetos modelados. Vamos a ver algunas de ellas que nos serán útiles (o inútiles)

### Listas

Podemos crear objetos a través de una "lista", que van a constar de una serie de vértices. 
Estos vértices estarán encerrados dentro de: 

    glNewList()
    #objetos
    glEndList()

Estas listas en general vienen con una identridad. Asignaremos 


```python
#Para el primer ejemplo
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys
import numpy as np

lados = 10

def init():
    """
    Función que inicializa la ventana
    """
    #Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    #El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    #El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0,-2.0,2.0)
    
def objetos():
    global fig
    #creamos lista
    fig = glGenLists(1)
    glNewList(fig, GL_COMPILE)
    glBegin(GL_POLYGON)
    
    
    for k in range(lados+1):
        glVertex(np.cos(2*np.pi/lados*k), np.sin(2*np.pi/lados*k))
    
    glEnd()
    glEndList()
    
    
def trianguloE():
    glClear(GL_COLOR_BUFFER_BIT)
    
    # Definir el color que tendrán los objetos que vamos a dibujar
    #glColor3f(1.0, 0.0, 0.0)
    #glLineWidth(2.0)
    #glBegin(GL_TRIANGLE_STRIP)
    #lados=6
    #for k in range(lados): #para cerrar el hexagono ya que solo encadena y llega hasta el lado 6 y se necesita un 7 vértice
    #    glVertex(np.cos(2*np.pi/lados*k), np.sin(2*np.pi/lados*k))
    #glEnd()

    glCallList(fig)#se llama a la lista y nos ahorramos lo anterior
    
    glFlush()
    
glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Probando figuras")# Creamos la ventana

init()# Inicializamos aquí a la ventana
objetos()

glutDisplayFunc(trianguloE)

glutMainLoop()
```
