```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

import numpy as np

# Librería PIL
from PIL import Image

# Cargamos la imagen
image = Image.open("box2.jpg")

# Guardamos las dimeniones de la imagen
ix, iy = image.size

# pasamos la imagen a bytes
textura = image.tobytes("raw", "RGBX", 0, -1)

# Tamaño de la ventana
xw, yw = 750, 750

xmin, xmax, ymin, ymax, dnear, dfar = -10, 10, -10, 10, 0.1, 100
#alpha = 40
#aspect = xw/yw
#dnear = 0.1
# dfar = 1000

def init():
    glClearColor(0.0, 0.0, 0.2, 1.0)
    
    glMatrixMode(GL_PROJECTION)
    
    glFrustum(xmin, xmax, ymin, ymax, dnear, dfar)
    # gluPerspective(alpha, aspect, dnear, dfar)
    print(glGetDouble(GL_PROJECTION_MATRIX))


def figure():
    global fig
    
    fig = glGenLists(1)
    

    quad  = [
        [0,0],
        [1,0],
        [1,1],
        [0,1]
    ]
    vertex = [
        [-5,-5,0],
        [5,-5,0],
        [5,5,0],
        [-5,5,0] 
    ]
    
    
    glPolygonMode(GL_FRONT_AND_BACK, GL_FILL)
    glTexParameterf(GL_TEXTURE_2D, GL_TEXTURE_MAG_FILTER, GL_NEAREST)
    glTexParameterf(GL_TEXTURE_2D, GL_TEXTURE_MIN_FILTER, GL_NEAREST)
    glTexEnvf(GL_TEXTURE_ENV, GL_TEXTURE_ENV_MODE, GL_DECAL)
    
    
    glTexImage2D(GL_TEXTURE_2D, 0, 3, ix, iy, 0, GL_RGBA, GL_UNSIGNED_BYTE, textura)
    
    # Comenzamos la lista llamada fig
    glNewList(fig, GL_COMPILE)

    glEnable(GL_TEXTURE_2D)
    glBegin(GL_QUADS)
    for i in range(4):
        glTexCoord2fv(quad[i])
        glVertex3fv(vertex[i])
    glEnd()
    glDisable(GL_TEXTURE_2D)
    
    glEndList()# Fin de la lista

def dibujo():

    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
    
    glMatrixMode(GL_MODELVIEW)
    
    glTranslatef(0,0,-5)
    glScalef(50, 50, 50)
    
    glCallList(fig)
    
    glFlush()

glutInit(sys.argv)

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB | GLUT_DEPTH)
glutInitWindowPosition(0, 0)
glutInitWindowSize(xw, yw)

glutCreateWindow(b"Plano")

init()
figure()# Cargamos nuestra figura

glutDisplayFunc(dibujo)
# glutIdleFunc(dibujo)# Con esta función, una vez que haya terminado de hacer un búfer, vuelve a ejecutar dibujo
glutMainLoop()
```
