Vamos a continuar dibujando objetos en el plano. 

## Segmentos

Para dibujar puntos usábamos

    glBegin(GL_POINTS)
    # Vértices
    glEnd()
Para dibujar segumento pues lo que usamos es justamente

    glBegin(GL_LINES)
    # Vértices
    glEnd()
    
Con esto, va tomando pares de puntos y dibuja el segmento que une a ambos vértices.

Existe otro comando para que *encadene* los puntos. Este comando es
    
    glBegin(GL_LINE_STRIP)
    # Vértices
    glEnd()

Existe otra versión más para dibujar líneas concatenadas, que justamente resolvería el problema del cerrar nuestro polígono

    glBegin(GL_LINE_LOOP)
    # Vértices
    glEnd()
Une el último de los vértices con el primero mediante un segmento. De esta manera, `GL_LINE_LOOP` determinará un polígono.

## Polígonos

La figura más básica para dibujar es el triángulo. ¿Por qué? Primero porque es el que tiene menos lados, pero (más importante) es que todo polígono se puede subdividir en triángulos. 

Los triángulos están definidos por sus tres vértices. Y en `OpenGL` podemos dibujar un triángulo a través de sus tres vértices con 
    
    glBegin(GL_TRIANGLES)
    # Vértices
    glEnd()
El conjunto de vértices que le damos como dato los agrupa de tres en tres y forma con cada grupo el triángulo que determinan esos tres vértices.

Si queremos realizar figuras un poco más complejas a partir de pegados de triángulos, este comando posiblemente nos haría repetir una y otra vez los vértices. Para esto tenemos otras dos versiones:

1. `GL_TRIANGLE_STRIP`: Si los vértices son $P_1,P_2,...,P_n$, realiza agrupaciones de tres en tres $P_k, P_{k+1}, P_{k+2}$ para $k=1,\ldots,n-2$
2. `GL_TRIANGLE_FAN`: Si los vértices son $P_0,P_1,P_2,...,P_n$, realiza las agrupaciones de tres en tres $P_0,P_k,P_{k+1}$ para $k=1,\ldots,n-1$, y dibuja el triángulo que determina cada una de las agrupaciones.

### Transparencias (básico)

Para poder añadir transparencias, tenemos que usar una cuarta coordenada que se conoce como *alpha*, y va  de 0 a 1 como el resto de las coordenadas: `0` es para *transparente* y `1` es para *opaco*.

Luego, en lugar de usar RGB, tendríamos que 

* usar RGBA: Añadimos `glutInitDisplayMode()` la constante `GLUT_RGBA`.
* Usar colores con cuatro componentes `glColor4f()`, donde la última coordenada es el *alpha*.
* tendríamos que habilitar las transparencias:

        glEnable(GL_BLEND)
y cuando activamos algo, podemos desactivarlo con 
        
        glDisable(GL_BLEND)
* y cómo se mezclan: esto se termina con la función `glBlendFunc()` que tiene dos argumentos.

        glBlendFunc(GL_SRC_ALPHA,GL_ONE_MINUS_SRC_ALPHA)


```python
### LIBRERÍAS
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

import numpy as np

def init():
    """
    Función que inicializa la ventana
    """
    # Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    # El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    # El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0, -2.0, 2.0)
    
def dibu():
    glClear(GL_COLOR_BUFFER_BIT)
    
    
    # Ancho de la línea
    glLineWidth(2.0)
    
    glEnable(GL_BLEND)
    glBlendFunc(GL_SRC_ALPHA, GL_ONE_MINUS_SRC_ALPHA)
    # Dibujamos un punto
    glBegin(GL_TRIANGLES)
    glColor4f(0.0, 0.0, 1.0, 0.25)# Azul
    glVertex2f(0,-2)
    glVertex2f(1,1)
    glVertex2f(-1,1)
    
    glColor4f(1.0, 0.0, 0.0,1)# Rojo
    glVertex2f(0,2)
    glVertex2f(1,-1)
    glVertex2f(-1,-1)
    

    glEnd()
    glDisable(GL_BLEND)
    glFlush()

glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGBA)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Lineas")# Creamos la ventana

init()# Inicializamos aquí a la ventana

glutDisplayFunc(dibu)

glutMainLoop()
```

## Listas

Podemos crear objetos a través de una "lista", que van a constar de una serie de vértices. Estos vértices estarán encerrados dentro de 
    
    glNewList()
    # Objeto
    glEndList()
Estas listas en general vienen con una identidad. Asignaremos esta identidad usando `glGenLists()`:
    
    nombre = glGenLists(1)
En `glNewList()` hay dos argumentos: el primero es el nombre del objeto que vamos a definir; y el segundo a elegir entre dos posibilidades: `GL_COMPILE` y `GL_COMPKILE_AND_EXECUTE`.

Para llamar a dicho objeto para que lo dibuje, tan sólo tenemos que escribir `glCallList(nombre)`

```python
### LIBRERÍAS
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

import numpy as np

lados = 10

def init():
    """
    Función que inicializa la ventana
    """
    # Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    # El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    # El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0, -2.0, 2.0)
    
def objetos():
    global fig, lados
    # Creamos la lista
    fig = glGenLists(1)
    # A continuación creamos la lista
    glNewList(fig, GL_COMPILE)
    glBegin(GL_POLYGON)
    for k in range(lados+1):
        glVertex2f(np.cos(2*np.pi/lados*k), np.sin(2*np.pi/lados*k))
    glEnd()
    glEndList()

def dibu():
    glClear(GL_COLOR_BUFFER_BIT)
    
    
    glColor3f(0.0, 0.0, 0.0)
    
    glCallList(fig)

    
    glFlush()

glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Lineas")# Creamos la ventana

init()# Inicializamos aquí a la ventana
objetos()

glutDisplayFunc(dibu)

glutMainLoop()
```

## Usar listas para describir objetos

Imáginaos que tengo una lista donde están todos los vértices (Que pueden repetirse) pero sin haber repetición. 

Vamos a tener una lista de vértices (que llamaremos `vertex`), y una lista de índices (que pueden repetirse) que llamaremos `index`:

Entonces usaremos la siguiente secuencoa

    glEnableClientState(GL_VERTEX_ARRAY)
    glVertexPointer(2, GL_FLOAT, 0, vertex)
    glDrawElements(GL_QUADS, len(vertex), GL_UNSIGNED_BYTE, index)
    glDisableClientState(GL_VERTEX_ARRAY)
El primer y el último comando habilita y deshabilita, respectivamente, la opción de describir un objeto mediante una pareja de vértices e índices.

El segundo comando indica cuál es el conjunto de vértices que vamos a utilizar. Tiene cuatro entradas

* Número de entradas de cada punto (aquí son 2 pero en general será 3)
* Tipo de dato (`GL_FLOAT`, `GL_DOUBLE`, etc)
* Desplazamiento (0 por defecto)
* nombre de la lista que contiene los vértices

En cuanto al tercer comando, indica qué queremos dibujar y cómo tenemos que recorrer los vértices:

* el tipo de dibujo (cualquiera de todos los que hemos visto antes)
* la longitud del array de vértices
* el tipo 

```python
### LIBRERÍAS
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

import numpy as np

lados = 10

# Lista de vértices
vertex = [ [1.,1.],
           [-1.,-1.],
           [1.,-1.],
          [-1.,1.]
         ]

# Lista de índices
index = [0, 2, 1, 3]

def init():
    """
    Función que inicializa la ventana
    """
    # Establecemos el color de fondo
    glClearColor(0.5, 0.5, 0.5, 0.0)
    
    # El tipo de proyección
    glMatrixMode(GL_PROJECTION)
    
    # El recorte que vamos a mostrar en la pantalla
    gluOrtho2D(-2.0, 2.0, -2.0, 2.0)
    


def dibu():
    global vertex
    glClear(GL_COLOR_BUFFER_BIT)
    
    
    glColor3f(0.0, 0.0, 0.0)
    
    glEnableClientState(GL_VERTEX_ARRAY)
    glVertexPointer(2, GL_FLOAT, 0, vertex)
    glDrawElements(GL_QUADS, len(vertex), GL_UNSIGNED_BYTE, index)
    glDisableClientState(GL_VERTEX_ARRAY)

    
    glFlush()

glutInit(sys.argv) # Inicializamos GLUT

glutInitDisplayMode(GLUT_SINGLE | GLUT_RGB)

glutInitWindowPosition(1000,500) # Posición de la ventana
glutInitWindowSize(500, 500)# Indica el tamaño de la ventana

glutCreateWindow("Lineas")# Creamos la ventana

init()# Inicializamos aquí a la ventana

glutDisplayFunc(dibu)

glutMainLoop()
```
